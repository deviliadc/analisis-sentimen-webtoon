# SCRAPING

```python
from google_play_scraper import reviews
import pandas as pd

# Mengambil ulasan dari Google Play
result, continuation_token = reviews(
    'com.naver.linewebtoon',  # ID aplikasi
    lang='id',           # Ulasan dalam bahasa Indonesia
    count=15000           # Jumlah ulasan yang ingin diambil
)

# Memproses hasil ulasan menjadi DataFrame pandas
reviews_data = pd.DataFrame(result)

# Menyimpan DataFrame ke file CSV
reviews_data.to_csv('/content/drive/MyDrive/Colab Notebooks/dicoding/ulasan.csv', index=False, encoding='utf-8')

print("Ulasan berhasil disimpan ke ulasan.csv")
```
